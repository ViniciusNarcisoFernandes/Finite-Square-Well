```python
# Equação transcendental par e ímpar

import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt

# Massa do elétron(kg)
m_e = 9.10938356 * 10**(-31)

# Constate de planck (eV.s):
hbar1 = 1.0545718 * 10**(-14)
hbar2 = 6.58211915 * 10**(-16)
hbar3 = hbar1*hbar2

# Potencial:
V = 10

# Tamanho do Poço:
a = 10

# Energia:
E = np.linspace(0,10,1000)

# Substituição:
z = np.sqrt((2*m_e*E)/hbar3)*(a/2)
z0 = np.sqrt((m_e*V*np.power(a,2))/(2*hbar3))

# Equação transcendental par:
def h1(z):
  return z*np.tan(z) - np.sqrt(np.power(z0,2) - np.power(z,2))

def f1(z):
  return z*np.tan(z)

def g1(z):
  return np.sqrt(np.power(z0,2) - np.power(z,2))

# Equação transcendental ímpar
def h2(z):
  return z/np.tan(z) + np.sqrt(np.power(z0,2) - np.power(z,2))

def f2(z):
  return z/np.tan(z)

def g2(z):
  return -np.sqrt(np.power(z0,2) - np.power(z,2))

# Pontos para a resolução da equação h1(z) e h2(z):
print("Ponto1 =",opt.fsolve(h1,1.25))
print("Ponto2 =",opt.fsolve(h2,1.50))
print("Ponto3 =",opt.fsolve(h1,1.75))
print("Ponto4 =",opt.fsolve(h2,3.50))
print("Ponto5 =",opt.fsolve(h1,4.75))
print("Ponto6 =",opt.fsolve(h2,7.00))

# Plot:
plt.figure(figsize=(10,6))
plt.plot(E,f1(E),color='Red', linestyle=':', label='f1(k2)')
plt.plot(E,g1(E),color='Blue', linestyle='-.', label='g1(k2)')
plt.plot(E,f2(E),color='Black', linestyle=':', label='f2(k2)')
plt.plot(E,g2(E),color='Green', linestyle='-.', label='g2(k2)')
plt.title('Poço Quadrado Finito: Estados Ligados')
plt.xlabel('E')
plt.ylabel('Pontos')
plt.ylim([-10,10])
plt.grid(True)
plt.show()
```

```python
# Função de onda: Ponto1

import numpy as np
import matplotlib.pyplot as plt

# Massa do elétron(kg):
m_e = 9.10938356 * 10**(-31)

# Constante de Planck(eV.s):
hbar1 = 1.0545718 * 10**(-14)
hbar2 = 6.58211915 * 10**(-16)
hbar3 = hbar1*hbar2

# Potencial:
V0 = 10

# Tamanho do poço:
a = 10

# Ponto 1 para equação transcendental par:
z1 = 1.39741765

# Escrevendo k's
k2 = z1/(a/2)
k1 = np.sqrt(((2*m_e*V0)/hbar3) - ((4*np.power(z1,2))/np.power(a,2)))

# Escrevendo as constantes C e D da função de onda normalizada
C = np.sqrt((2*k1)/(2*np.power(np.sin((k2*(a/2))),2) + k1*a))
D = np.sqrt((2*k1)/(2*np.power(np.cos((k2*(a/2))),2) + k1*a))

# Função de onda quando C = A - G = 0 (Equação transcendental par), Temos:
def psi1(x):
  return D*np.cos(k2*(a/2))*np.exp(k1*(a/2))*np.exp(k1*x)

def psi2(x):
  return D*np.cos(k2*x)

def psi3(x):
  return D*np.cos(k2*(a/2))*np.exp(k1*(a/2))*np.exp(-k1*x)

# Usando os limites:
x1 = np.arange(-10 , -5 , 0.01)
x2 = np.arange(-5 , 5 , 0.01)
x3 = np.arange(5 , 10 , 0.01)

# Plot:
plt.figure(figsize=(10,6))
plt.plot(x1,psi1(x1), color ='black', linestyle = '-.', label = '$x_1$')
plt.plot(x2,psi2(x2), color ='red', linestyle = '-.', label = '$x_2$')
plt.plot(x3,psi3(x3), color ='Orange', linestyle = '-.', label = '$x_3$')
plt.xlim([-11,11])
plt.ylim([-0.10,0.5])
plt.title('Função de Onda: Ponto1')
plt.xlabel('x')
plt.ylabel('$\psi_{1}$(x)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
