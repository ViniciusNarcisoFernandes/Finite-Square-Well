```python
import numpy as np
import matplotlib.pyplot as plt

def Numerov_left(a,E,U,N,xm,xll,C):

  # Valor inicial xleft:
  xl = x1 = a
  xll = xll
  xmatch = xm

  # Substituição:
  h1 = (xmatch-x1)/N
  b1 = np.power(h1,2)/12

  # Número de onda:
  kl = np.zeros([N])
  k1 = np.sqrt((2*m_e*(U-E))/hbar3)
  kl[0] = -np.power(k1,2)
  kl[1] = -np.power(k1,2)

  # Função de onda:
  psil = np.zeros([N])
  psil[0] = C*np.exp(k1*x1)
  psil[1] = C*np.exp(k1*(x1+h1))

  # Numerov:
  for n in range(1,N-1):
    xl += h1
    if (xl > (xll)):
      Ul = 0
    else:
      Ul = U
    k1 = (2*m_e*(E-Ul))/hbar3
    kl[n+1] = k1
    psil[n+1] = ((2*(1-(5*b1*kl[n]))*psil[n]) - ((1+(b1*kl[n-1]))*psil[n-1]))/(1+(b1*kl[n+1]))

  return psil

def Numerov_right(b,E,U,N,xm,xrr,D):

  # Valor inicial xleft:
  xr = x2 = b
  xrr = xrr
  xmatch = xm

  # Substituição :
  h2 = (x2-xmatch)/N
  b2 = np.power(h2,2)/12

  # Número de onda:
  kr = np.zeros([N])
  k2 = np.sqrt((2*m_e*(U-E))/hbar3)
  kr[0] = -np.power(k2,2)
  kr[1] = -np.power(k2,2)

  # Função de onda:
  psir = np.zeros([N])
  psir[0] = D*np.exp(-k2*x2)
  psir[1] = D*np.exp(-k2*(x2-h2))

  # Numerov:
  for n in range(1,N-1):
    xr -= h2
    if (xr < (xrr)):
      Ur = 0
    else:
      Ur = U
    k2 = (2*m_e*(E-Ur))/hbar3
    kr[n+1] = k2
    psir[n+1] = ((2*(1-(5*b2*kr[n]))*psir[n]) - ((1+(b2*kr[n-1]))*psir[n-1]))/(1+(b2*kr[n+1]))

  return psir

# Constantes:
m_e = 9.10938*10**(-31)
hbar1 = 1.0545718 * 10**(-14)#(kg.m^2/s)
hbar2 = 6.58211915 * 10**(-16)#(eV.s)
hbar3 = hbar1*hbar2

# Pontos:
a = int(input('a: '))
b = int(input('b: '))
xm = int(input('xmatch: '))
xll = int(input('Xll: '))
xrr = int(input('Xrr: '))

# Energia:
Ei = 0
Ef = int(input('Ef: '))

# Potencial:
U = int(input('U: '))

# Número de pontos:
N = 1000

# Valor de h:
h = (b-xm)/N

# Lista:
Res1 = list() ; Res2 = list()

# Energia:
for i in np.arange(Ei,Ef+h,h):
  E = i
  # Função left:
  Numl = Numerov_left(a,E,U,N,xm,xll,C)
  for i,j in enumerate(Numl):
    if (i == len(Numl)-1):
      l1 = j
    elif (i == len(Numl)-2):
      l2 = j
    elif (i == len(Numl)-3):
      l3 = j

  left = (((1/2)*(l1)) - (2*l2) + ((3/2)*(l3)))/h

  # Função right:
  Numr = Numerov_right(b,E,U,N,xm,xrr,D)
  for i,j in enumerate(Numr):
    if (i == len(Numr)-1):
      r1 = j
    elif (i == len(Numr)-2):
      r2 = j
    elif (i == len(Numr)-3):
      r3 = j

  right = ((-(3/2)*(r3)) + (2*r2) - ((1/2)*(r1)))/h

  # Soma  dos pontos:
  R1 = l1-r1
  Res1.append(R1)
  # Soma das derivadas:
  R2 = left - right
  Res2.append(R2)

# Plot:
plt.figure()
plt.plot(Res1,color='Purple',linestyle='--',label='Psi_left')
plt.plot(Res2,color='Black',linestyle='--',label='Psi_right')
plt.title('Numerov')
plt.xlabe('N')
plt.ylabel('Soma dos pontos + derivadas')
plt.ylim([-0.015,0.005])
plt.grid(True)
plt.show()
```
